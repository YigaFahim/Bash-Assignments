```python
import plotly.express as px
```

```python
ag3 = malariagen_data.Ag3()
```

```python
ag3_me = ag3.sample_metadata(sample_sets="3.0")
ag3_me.query("country == 'Ghana'")
```

```python
Ghana_data = ag3_me.query("country == 'Ghana'")
Ghana_data
```

```python
Ghana_data.groupby("location").size()
```

```python
Ghana_data.groupby("year").size()
```

```python
pivot_country_taxon = (
    Ghana_data
    .pivot_table(
        index="location",
        columns="taxon",
        values="sample_id",
        aggfunc="count",
        fill_value=0
    )
)
pivot_country_taxon
```

```python
fig = px.bar(pivot_country_taxon, height=600, width=800)
fig.update_layout(
    title="Ag3.0 genomes sequenced",
    yaxis_title="no. genomes",
)
fig
```

```python
pivot_location_taxon = (
    Ghana_data
    .pivot_table(
        index=["location", "latitude", "longitude"],
        columns=["taxon"],
        values="sample_id",
        aggfunc="count",
        fill_value=0,
    )
)

pivot_location_taxon
```

```python
import ipyleaflet
```

```python
# create a map
m = ipyleaflet.Map(
    basemap=ipyleaflet.basemaps.OpenStreetMap.Mapnik,
    center=[8, 0],
    zoom=6.3,
)

# add markers for sampling locations
for row in pivot_location_taxon.reset_index().itertuples():
    marker = ipyleaflet.Marker(
        location=(row.latitude, row.longitude),
        draggable=False,
    )
    m.add_layer(marker)

# add a scale bar
m.add_control(ipyleaflet.ScaleControl(position="bottomleft"))

# display the map
m
```
